```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
```

```python
#  CSV file into the df
crypto_df = pd.read_csv('crypto_data.csv')
crypto_df.head()
```

```python
#  Shape
crypto_df.shape
```

```python
crypto_df.value_counts()
```

```python
# only currencies that are currently being traded
crypto_df = crypto_df[crypto_df['IsTrading'] == True]
crypto_df.head()
```

```python
# drop [IsTrading] column
crypto_df = crypto_df.drop(columns ='IsTrading')
```

```python
# check for null values
crypto_df.isnull().sum()
```

```python
# remove null values
crypto_df = crypto_df.dropna()
```

```python
# make sure nulls are dropped
crypto_df.isnull().sum()
```

```python
# filter cryptocurrencies that have been mined
crypto_df = crypto_df[crypto_df["TotalCoinsMined"]> 0]
crypto_df
```

```python
# Since the coin names do not contribute to the analysis of the data, 
# delete the CoinName from the dataframe.
crypto_df.drop(columns='CoinName', axis=1, inplace= True)
crypto_df
```

```python
crypto_df = crypto_df.drop(columns=['Unnamed: 0'])
crypto_df
```

```python
# convert Algorithm and ProofType into numerical data
X = pd.get_dummies(crypto_df, columns=['Algorithm', 'ProofType'])
X
```

```python
# column names 
X.columns
```

```python
# standarize dataset
scale = StandardScaler()
crypto_scale = scale.fit_transform(X)
```

```python
# number of features
len(X.columns)
```

```python
#  create a model that will preserve approximately 99% of the explained variance
pca_variance = PCA(n_components=.90)

# Get three principal components for the crypto data.
crypto_pca = pca_variance.fit_transform(crypto_scale)
crypto_pca
```

```python
# create DataFrame
crypto_pca_trans = pd.DataFrame(data=crypto_pca)
crypto_pca_trans.head()
```

```python
pca.explained_variance_ratio_
```

```python
# t-SNE model
tsne = TSNE(learning_rate=35)
```

```python
# Reduce dimensionsa and check that dimensions have been reduced to two
tsne_features = tsne.fit_transform(crypto_pca_trans)
tsne_features.shape
```

```python
# The first column of transformed features
x = tsne_features[:,0]

# The second column of transformed features
y = tsne_features[:,1]
```

```python
plt.scatter(x, y)
plt.show()
```
